#### Removing the Constant Feature

###### Constant features are those features which have same value for all rows/observation  of dataset

###### We are going to use VarianceThreshold function from sklearn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### VarianceTHreshold Function

###### variance threshold from sklearn is the simple baseline approach of feature selection where it removes those features


###### which variance doesn't meet certain threshold,by default it remove all zero variance features

```python
from sklearn.feature_selection import VarianceThreshold
model = VarianceThreshold(threshold=0)
model.fit(X_train)
```

```python
model.fit(X_test)
```

###### we use get_support function to know how many functions are not constant

```python
sum(model.get_support())
```

##### Now reducing  training and testing sets using transform function this process remove all the constant features and change the dimension of the training and testing dataset

```python
X_train = model.transform(X_train)
X_test = model.transform(X_test)
```

# Removing Quasi-Constant Features

###### quasi-constant features are those features that show same value for the great majority of the observations of the dataset and doesn't provide any sort of information that the machine learning model can utilize that's why we remove this sort of features using method called VarianceThreshold function from sklearn

###### 

###### while removing the quasi constant features we just set the threshold value(threshold=0.01) for VarianceThreshold

# Information Gain 

###### Information gain calculate the average/sum of the entropy of the split in Decision Tree from top to bottom of the split. The higher the information gain of the certain split the better is that split.

According to Google Information Gain is actually a method that measures how much information the presenece/absence of a features contribute to make the correct prediction on the target
